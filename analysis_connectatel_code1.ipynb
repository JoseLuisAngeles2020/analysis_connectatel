-El comportamiento de uso entre edades fue bastante parecido. En promedio, los tres grupos enviaron cerca de 5.5 mensajes, realizaron alrededor de 4.5 llamadas y consumieron entre 22.7 y 23.1 minutos.

Esto sugiere que la edad, por sí sola, no marca una diferencia fuerte en el consumo, por lo que no parece ser la mejor variable para definir planes muy distintos.

📊 Segmentos por Nivel de Uso

-El segmento dominante fue Uso medio, con 2943 usuarios (73.58%).

-Después apareció Bajo uso, con 779 usuarios (19.48%).

-El grupo más pequeño fue Alto uso, con 278 usuarios (6.95%).

La composición por edad dentro de estos grupos también fue muy estable:

-En Adultos, 74.33% están en uso medio.

-En Adultos Mayores, 72.91% están en uso medio.

-En Jóvenes, 72.63% están en uso medio.

En términos de volumen, el grupo de Adultos es el más importante para el negocio porque combina mayor tamaño y también la mayor cantidad absoluta de usuarios de alto uso.

💼 Segmentos más valiosos para ConnectaTel

El segmento más valioso en escala es Uso medio, porque concentra casi 3 de cada 4 clientes. Es el mejor grupo para estrategias de retención, upgrades y venta de paquetes adicionales.

📈 Patrones de uso extremo (outliers)

-cant_mensajes: 46 usuarios (1.15%)

-cant_llamadas: 30 usuarios (0.75%)

-cant_minutos_llamada: 96 usuarios (2.4%)

Estos outliers no son errores de captura de datos, son datos excepcionales donde los usuarios tuvieron un consumo alto a comparación del promedio.

💡 Recomendaciones

-Crear una oferta clara para usuarios de alto uso, especialmente orientada a quienes acumulan muchos minutos de llamada.

-Mantener un plan simple y económico para usuarios de bajo uso, evitando que paguen por beneficios que no aprovechan.

-Rediseñar los planes y compportamientos para evitar disrupciones entre aquellos usuarios de alto y bajo consumo.
